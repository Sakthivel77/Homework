## Exercise 6.4

Write a function that takes in a string, `s`, and returns a reversed copy of `s`. The only string operation you are allowed to use is string concatenation.

>```
>your_function(`CJN`)
>NJC
>```

````python
def reverze(a):
    x = ''
    for i in range(1,len(a)+1):
        x += a[-i]
    return x
        
def reverze_recurze(a):
    if a == '':
        return a
    else:
         return a[-1]+ reverze_recurze(a[:-1])
print(reverze_recurze('CJN'))
print(reverze('CJN'))
````
