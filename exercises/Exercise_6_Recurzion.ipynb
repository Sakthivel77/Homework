## Exercise 6.1 

Write a function that takes in a positive integer $n$, and prints the numbers from $n$ to $0$ (in descending order).
Example interaction: 

>```
>your_function(3)
>3
>2
>1
>0
>```

````python
def your_function(n):
    if n == 0:
        print(0)
    else:
        print(n)
        your_function(n-1)


def my_function(n):
    for i in range(n+1)[::-1]:
        print(i)
my_function(5)
````

## Exercise 6.4

Write a function that takes in a string, `s`, and returns a reversed copy of `s`. The only string operation you are allowed to use is string concatenation.

>```
>your_function(`CJN`)

>```

````python
def reverze(a):
    x = ''
    for i in range(1,len(a)+1):
        x += a[-i]
    return x
        
def reverze_recurze(a):
    if a == '':
        return a
    else:
         return a[-1]+ reverze_recurze(a[:-1])
print(reverze_recurze('CJN'))
print(reverze('CJN'))
````
